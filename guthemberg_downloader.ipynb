```python
import os
import csv
import random
import requests

url_template = "http://www.gutenberg.org/cache/epub/{0}/pg{0}{1}"

def download_book(book_id, title, directory):
    for ext in [".txt", "-0.txt"]:
        url = url_template.format(book_id, ext)
        response = requests.get(url)
        if response.status_code == 200:
            if not os.path.exists(directory):
                os.makedirs(directory)
            filename = os.path.join(directory, f"{title}.txt")
            filename = filename.replace("\n","")[:70] #remove invalid chars and too long titles
            with open(filename, "wb") as f:
                f.write(response.content)
            return True
    return False

def download_random_books(csv_path, num_books=500, genre="Fiction", language="en"):
    book_ids = []
    with open(csv_path, mode="r", encoding="utf-8") as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            book_id = row["Text#"]
            title = row["Title"].replace('"', '').replace('/', '-')
            subjects = row["Subjects"]
            lang = row["Language"]
            if genre in subjects and lang == language:
                book_ids.append((book_id, title))

    random.shuffle(book_ids)

    downloaded_books = 0
    subdirectory = "books"
    for book_id, title in book_ids:
        if download_book(book_id, title, subdirectory):
            downloaded_books += 1
            print(f"Downloaded {downloaded_books}: {title}")
            if downloaded_books >= num_books:
                break

csv_path = "./pg_catalog.csv"
download_random_books(csv_path)
```
